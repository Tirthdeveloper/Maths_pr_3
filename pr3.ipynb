```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Copy of research_student (1).csv")
df.head()
```

```python
from scipy import stats
from statsmodels.stats.weightstats import zconfint
from scipy.stats import ttest_ind, chi2_contingency, f_oneway
import pandas as pd
import numpy as np


numerical_cols = [
    'CGPA',
    'Rank',
    'GPA 1',
    'GPA 2',
    'Technical Projects',
    'Tech Quiz'
]

print("95% Confidence Interval\n")

for col in numerical_cols:

    # Remove missing values
    data = df[col].dropna()

    # Confidence Interval
    ci = zconfint(data)

    print(f"{col}: {ci}")
```

```python
from scipy import stats

# Significance level
alpha = 0.05

# Critical value for two-tailed test
critical_value = stats.norm.ppf(1 - alpha/2)

print("Critical Value:", critical_value)
print("Significance Level:", alpha)
```

```python
from scipy.stats import ttest_ind

# Create two groups
group1 = df[df['Technical Projects'] >= 5]['CGPA']
group2 = df[df['Technical Projects'] < 5]['CGPA']

# Perform T-Test
t_stat, p_value = ttest_ind(group1, group2)

print("\nT-Test Result")
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Decision
if p_value < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
# 5. Chi-Square Test

import pandas as pd
from scipy.stats import chi2_contingency

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Create category for CGPA
df['CGPA_Category'] = pd.cut(
    df['CGPA'],
    bins=[0, 7, 10],
    labels=['Low', 'High']
)

# Contingency Table
contingency_table = pd.crosstab(
    df['CGPA_Group'],
    df['CGPA_Category']
)

# Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Test")
print("Chi-Square Value:", chi2)
print("P-Value:", p)

# Decision
if p < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
# 6. ANOVA Test

from scipy.stats import f_oneway
import pandas as pd

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Create CGPA Groups
df['CGPA_Group'] = pd.cut(
    df['CGPA'],
    bins=[0, 6, 7, 8, 10],
    labels=['Below Average', 'Average', 'Good', 'Excellent']
)

# Groups based on CGPA category
group1 = df[df['CGPA_Group'] == 'Below Average']['Tech Quiz']
group2 = df[df['CGPA_Group'] == 'Average']['Tech Quiz']
group3 = df[df['CGPA_Group'] == 'Good']['Tech Quiz']
group4 = df[df['CGPA_Group'] == 'Excellent']['Tech Quiz']

# ANOVA Test
anova_result = f_oneway(group1, group2, group3, group4)

print("\nANOVA Test")
print("F-Statistic:", anova_result.statistic)
print("P-Value:", anova_result.pvalue)

# Decision
if anova_result.pvalue < 0.05:
    print("Reject H0")
else:
    print("Accept H0")
```

```python
# 7. Covariance and Correlation

import numpy as np

# Remove missing values
data = df[['CGPA', 'Tech Quiz']].dropna()

# Covariance
covariance = np.cov(
    data['CGPA'],
    data['Tech Quiz']
)[0][1]

# Correlation
correlation = np.corrcoef(
    data['CGPA'],
    data['Tech Quiz']
)[0][1]

print("\nCovariance:", covariance)
print("Correlation:", correlation)
```
